## Ejemplo de Jardinero

```python
# --- ASISTENTE IA: JARDINERO Y PODADORA ---

print("==========================================")
print("   BIENVENIDO AL SISTEMA DE JARDINERÍA IA   ")
print("==========================================")

# Variable de control para el menú
opcion = 0

# Estructura repetitiva para mantener el programa activo
while opcion != 3:
    print("\n--- MENÚ PRINCIPAL ---")
    print("1. Diagnosticar necesidad de podado")
    print("2. Verificar estado de la podadora")
    print("3. Apagar sistema")

    # Captura de datos interactiva
    opcion = int(input("\nSeleccione una opción (1-3): "))

    # Lógica de "Inteligencia Artificial" mediante condicionales
    if opcion == 1:
        print("\n--- ESCANEO DEL JARDÍN ---")
        altura_pasto = float(input("Ingrese la altura del pasto en cm: "))
        clima = input("¿Cuál es el clima actual? (sol/lluvia): ")
        obstaculos = input("¿Hay objetos en el jardín? (si/no): ")

        # Razonamiento lógico basado en múltiples criterios y operadores
        if clima == "lluvia":
            print("IA DECISIÓN: OPERACIÓN CANCELADA.")
            print("Motivo: El pasto mojado daña las cuchillas y el suelo.")
        elif altura_pasto > 10 and obstaculos == "no":
            print("IA DECISIÓN: INICIAR PODADO AUTOMÁTICO.")
            print("Sugerencia: El pasto está alto y el área está despejada.")
        elif obstaculos == "si":
            print("IA ALERTA: PELIGRO DETECTADO.")
            print("Acción: Retire los objetos antes de que la IA active las cuchillas.")
        else:
            print("IA DECISIÓN: ESTADO ÓPTIMO.")
            print("Sugerencia: El pasto tiene una altura estética adecuada.")

    elif opcion == 2:
        print("\n--- ESTADO DE LA PODADORA ---")
        nivel_bateria = int(input("Ingrese nivel de batería (0-100): "))
        filo_cuchilla = int(input("Ingrese nivel de filo (1-10): "))

        # Evaluación de rendimiento mediante operadores relacionales
        if nivel_bateria < 20:
            print("IA SUGERENCIA: Regresar a la base de carga inmediatamente.")
        elif filo_cuchilla < 4:
            print("IA SUGERENCIA: Requiere mantenimiento. Cuchillas desafiladas.")
        elif nivel_bateria > 80 and filo_cuchilla > 8:
            print("IA ESTADO: Podadora al 100% de eficiencia para trabajo pesado.")
        else:
            print("IA ESTADO: Equipo listo para trabajo ligero.")

    elif opcion == 3:
        print("\nCerrando Asistente de Jardinería. ¡Hasta pronto, jardinero!")

    else:
        # Manejo de entradas no válidas
        print("Opción no válida. Intente de nuevo con un número del 1 al 3.")

print("------------------------------------------")
print("Simulación de IA de Obra finalizada.")

print("Programa realizado por Cepeda Darrell NL = 10")
```
